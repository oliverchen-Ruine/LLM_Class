## 使用qwen3筛选关键实验参数

```python
import requests

def query_ollama(prompt, model="qwen3:14b"):
    response = requests.post(
        "http://localhost:11434/api/generate",
        json={
            "model": model,
            "prompt": prompt,
            "stream": False
        }
    )
    return response.json()["response"]

query = """
我正在研发一种微胶囊化天然驱蚊精油的防蚊纤维材料。请帮我分析实验中的关键参数，并按重要性排序。

材料组成：
- 核材：香茅油(60%)+柠檬桉油(30%)+薄荷醇(10%)
- 壁材：改性壳聚糖(70%)+阿拉伯胶(30%)
- 基础纤维：改性聚酯(PET)

制备工艺：
1. 微胶囊制备：乳化-交联法
2. 纤维制备：熔融纺丝
3. 功能整理：浸渍-轧光-烘干

请列出每个环节中最关键的3-5个参数，给出合理的取值范围，并解释这些参数对最终产品性能的影响。
"""

print(query_ollama(query))
```

## 使用deepseek-r1优化实验流程

```python
import requests

def query_ollama(prompt, model="deepseek-r1:7b"):
    response = requests.post(
        "http://localhost:11434/api/generate",
        json={
            "model": model,
            "prompt": prompt,
            "stream": False
        }
    )
    return response.json()["response"]

query = """
我正在进行微胶囊化防蚊纤维材料的实验，已经确定了最优参数组合，现在需要优化实验流程以提高效率和可重复性。请帮我优化以下实验流程，指出可能的瓶颈和改进方案。

当前实验流程：
1. 微胶囊制备
   - 准备壁材溶液：壳聚糖溶于1%乙酸溶液，阿拉伯胶溶于去离子水
   - 准备核材溶液：将香茅油、柠檬桉油和薄荷醇混合
   - 乳化：将核材加入壁材溶液中，高速剪切乳化
   - 交联：加入戊二醛，控制pH和温度，搅拌4小时
   - 分离：离心分离微胶囊
   - 洗涤：用去离子水洗涤3次
   - 干燥：冷冻干燥12小时

2. 纤维制备
   - 聚合物准备：将PET切片干燥4小时
   - 混合：将微胶囊与PET切片混合
   - 熔融纺丝：设定温度、速度等参数
   - 牵伸：设定牵伸比
   - 卷绕：收集纤维

3. 性能测试
   - 包封率测定：溶剂提取法
   - 防蚊效果测试：WHO锥形测试法
   - 耐洗性测试：标准洗涤20次后重测防蚊效果
   - 纤维强度测试：拉伸测试

请从时间效率、资源利用、质量控制和可重复性四个方面提出优化建议。
"""

print(query_ollama(query))
```
